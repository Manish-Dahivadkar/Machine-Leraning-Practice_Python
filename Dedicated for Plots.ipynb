# Qutlier by boxplot or perctile function

```python
import pandas as pd
```

```python
import numpy as np
```

```python
prop = pd.read_csv(r"C:\Users\Manish\Desktop\R-classroom\datasets\Property_Price_Train.csv")
```

```python
def remove_outliers_boxplot(df, col):
    q3 = np.quantile(df[col],0.75)
    q1 = np.quantile(df[col], 0.25)
    iqr = q3-q1
    print ("iqr is ....", iqr)
    global outlier_higher_side
    global outlier_lower_side
    outlier_higher_side = [x for x in df[col] if (x >(q3+1.5*iqr))]
    outlier_lower_side = [x for x in df[col] if (x < (q1-1.5*iqr))]
```

```python
remove_outliers_boxplot(prop, "Sale_Price")
```
